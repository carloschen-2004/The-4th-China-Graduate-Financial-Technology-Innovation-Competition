### C类产品-客户聚类

```python
import pandas as pd
```

```python
df_C = pd.read_csv("./event_cust_C.csv")
df_C = df_C[df_C['first_cluster_id'] != 'C0']
df_C = df_C.dropna(axis=1, how='all')
df_C.head()
```

```python
df_C["first_cluster_id"].value_counts()
```

**特征处理：1.归一化 2. 缺失值赋0**

```python
cols_to_normalize = ["event_term","event_rate","event_amt","edu_bg","marriage_situ_cd",]
df_C[cols_to_normalize] = df_C[cols_to_normalize].apply(lambda x: ((x - x.mean()) / x.std()).fillna(0))
```
